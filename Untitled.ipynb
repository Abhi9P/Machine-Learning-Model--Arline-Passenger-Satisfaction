## Handling of Missing Data

```python
# Missing data
total = train.isnull().sum().sort_values(ascending=False)
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing.head()
```

```python
from sklearn.impute import SimpleImputer

# Create an instance of SimpleImputer with the mean strategy for numerical columns
imputer_num = SimpleImputer(strategy="mean")

# Create an instance of SimpleImputer with the most frequent strategy for categorical columns
imputer_cat = SimpleImputer(strategy="most_frequent")

# Loop through columns and apply imputation for both train and test datasets
for column in train.columns:
    if train[column].dtype in ['int64', 'float64']:
        train[column] = imputer_num.fit_transform(train[[column]])
        test[column] = imputer_num.transform(test[[column]])
    else:
        train[column] = imputer_cat.fit_transform(train[[column]])
        test[column] = imputer_cat.transform(test[[column]])
```

```python
# rechecking for missing value
total = train.isnull().sum().sort_values(ascending=False)
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing.head()
```

## Outliers Detection and Removal

```python
Q1 = train.quantile(0.25)
Q3 = train.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Removing outliers from dataset
train = train[~((train < (Q1 - 1.5 * IQR)) |(train > (Q3 + 1.5 * IQR))).any(axis=1)]
train.shape
```

```python
import time
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, plot_confusion_matrix, plot_roc_curve
from matplotlib import pyplot as plt 
def run_model(model, X_train, y_train, X_test, y_test, verbose=True):
    t0=time.time()
    if verbose == False:
        model.fit(X_train,y_train.ravel(), verbose=0)
    else:
        model.fit(X_train,y_train.ravel())
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred) 
    time_taken = time.time()-t0
    print("Accuracy = {}".format(accuracy))
    print("ROC Area under Curve = {}".format(roc_auc))
    print("Time taken = {}".format(time_taken))
    print(classification_report(y_test,y_pred,digits=5))
    plot_confusion_matrix(model, X_test, y_test,cmap=plt.cm.pink, normalize = 'all')
    plot_roc_curve(model, X_test, y_test)                     
    
    return model, accuracy, roc_auc, time_taken
```

```python
#Scaling the features with pipeline method, and standard scaler
pipeline = Pipeline([
            ('std_scaler',StandardScaler()),
                    ])
scaled_X_train = pipeline.fit_transform(X_train)
scaled_X_test = pipeline.transform(X_test)
```

```python
kfold = StratifiedKFold(n_splits=5)
```

```python
# Modeling step to test differents algorithms 
random_state = 2
classifiers = []
classifiers.append(LogisticRegression(random_state = random_state))
classifiers.append(KNeighborsClassifier())
classifiers.append(SVC(random_state=random_state))
classifiers.append(DecisionTreeClassifier(random_state=random_state))
classifiers.append(RandomForestClassifier(random_state=random_state))
classifiers.append(GradientBoostingClassifier(random_state=random_state))
classifiers.append(XGBClassifier(random_state=random_state))

cv_results = []
for classifier in classifiers :
    cv_results.append(cross_val_score(classifier, scaled_X_train, y = y_train, scoring = "accuracy", cv = kfold, n_jobs=4))

cv_means = []
cv_std = []
for cv_result in cv_results:
    cv_means.append(cv_result.mean())
    cv_std.append(cv_result.std())

cv_res = pd.DataFrame({"CrossValMeans":cv_means,"CrossValerrors": cv_std,"Algorithm":["LogisticRegression","KNeighbors","SVC","DecisionTree",
"RandomForest","GradientBoosting","XGBoost"]})

g = sns.barplot("CrossValMeans","Algorithm",data = cv_res, palette="Set3",orient = "h",**{'xerr':cv_std})
g.set_xlabel("Mean Accuracy")
g = g.set_title("Cross validation scores")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

# Create the RandomForestClassifier model with your specified parameters
model_rf = RandomForestClassifier(**params_rf)

# Define the number of training samples to use for the learning curve
train_sizes = np.linspace(0.1, 1.0, 10)  # You can adjust the range and number of values

# Calculate learning curves
train_sizes, train_scores, test_scores = learning_curve(
    model_rf, X_train, y_train, cv=5, train_sizes=train_sizes, scoring='accuracy', n_jobs=-1)

# Calculate the mean and standard deviation of scores
train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

# Plot the learning curve
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, label='Train Accuracy')
plt.plot(train_sizes, test_scores_mean, label='Validation Accuracy')
plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.2)
plt.xlabel('Number of Training Samples')
plt.ylabel('Accuracy')
plt.title('Learning Curve for RandomForestClassifier')
plt.legend()
plt.grid()
plt.show()
```

- the training accuracy increases rapidly, indicating that the model is quickly learning from the data. The fact that the line starts closer to 0 on the y-axis suggests that the model's initial accuracy is not very high, but it improves rapidly with more data.
- As the number of training samples increases beyond 10,000, the training accuracy continues to improve, albeit at a slower rate. This suggests that the model is still gaining insights from the additional data.
- after reaching around 50,000 training samples, the improvement in training accuracy starts to slow down significantly. This indicates that the model has captured most of the relevant patterns in the data, and adding more data has limited impact on further improving training accuracy.
- the initial low accuracy, even with a larger number of samples, might indicate a limitation in the model's complexity or architecture. It could also suggest that the data is inherently difficult to model accurately, leading to a high bias situation.
- The initial low accuracy could also be a result of noisy or insufficient data. Ensuring data quality and exploring possible sources of error is important.
